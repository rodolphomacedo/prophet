```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
```

Iremos ver nessa parte:

    - Usando dados mensais
    - Usando dados intra-di√°rios
    - Usando dados com Gap regulares

```python
df = pd.read_csv('./data/AirPassengers.csv')
df['Month'] = pd.to_datetime(df['Month'])
df.columns = ['ds', 'y']
df.head()
```

```python
model = Prophet(seasonality_mode='multiplicative')
model.fit(df)
```

```python
future = model.make_future_dataframe(periods=365*5)
forecast = model.predict(future)
fig = model.plot(forecast, xlabel='Date', ylabel=r'y data')
plt.title('Air Passagenrs - Errado - Dados Futuros n√£o d√°rios')
plt.show()
```

Os dados s√£o mensais, os dados futuros s√£o di√°rios.

```python
forecast.ds.tail()
```

```python
future = model.make_future_dataframe(periods=12*5, freq='MS')
forecast = model.predict(future)
fig = model.plot(forecast, xlabel='Date', ylabel=r'y data')
plt.title('Air Passagenrs - Certo - Dados Mensais')
plt.show()
```

# Dados Sub-di√°rios

```python
df = pd.read_csv('./data/divvy_hourly.csv')
df['date'] = pd.to_datetime(df['date'])
df.columns = ['ds', 'y']
df.head()
```

```python
model = Prophet(seasonality_mode='multiplicative')
model.fit(df)

future = model.make_future_dataframe(
    periods=365 * 24,
    freq='h'
)
```

```python
forecast = model.predict(future)
fig = model.plot(forecast)

plt.show()

fig2 = model.plot_components(forecast)
plt.show()
```

## Dados com gaps regulares

```python
df = df[(df['ds'].dt.hour >= 8) & (df['ds'].dt.hour < 18)]
```

```python
plt.plot(df['y'], 'o', ms=1)
plt.show()
```

```python
model = Prophet(seasonality_mode='multiplicative')
model.fit(df)
future = model.make_future_dataframe(periods=365*24, freq='h')

forecast = model.predict(future)
fig = model.plot(forecast)
plt.show()
```

Esse √© o mesmo problema anterior, estamos fazendo a predi√ß√£o para dias que n√£o temos a estimativa do Prophet.

```python
fig = model.plot(forecast)
plt.xlim(pd.to_datetime(['2018-08-01', '2018-08-04']))
plt.ylim(-2000, 4000)
plt.show()
```

Construindo a estimativa apenas no per√≠odo entre $8$ e $18$

```python
future2 = future[(future['ds'].dt.hour >= 8) & (future['ds'].dt.hour < 18)]
forecast2 = model.predict(future2)
fig = model.plot(forecast2)
plt.show()
```

```python
fig = model.plot(forecast2)
plt.xlim(pd.to_datetime(['2018-08-01',  '2018-08-04']))
plt.ylim(-2000, 4000)
plt.show()
```

Perceba que n√£o existe previs√£o para hor√°rios que n√£o temos estimativas.

```python
model.plot_components(forecast2)
plt.show()
```
